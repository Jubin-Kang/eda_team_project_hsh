```python
data = interest_rate.iloc[0:5, 1:]


averages = data.mean(axis=1, skipna=True)
max_values = data.max(axis=1, skipna=True)
min_values = data.min(axis=1, skipna=True)

max_banks = data.idxmax(axis=1, skipna=True)
min_banks = data.idxmin(axis=1, skipna=True)

summary = pd.DataFrame({
    'Average': averages,
    'Max Value': max_values,
    'Max Bank': max_banks,
    'Min Value': min_values,
    'Min Bank': min_banks
})
summary.index = ['2020년','2021년','2022년','2023년','2024년']
summary
```

```python
averages = data.mean(axis=1, skipna=True)
max_values = data.max(axis=1, skipna=True)
min_values = data.min(axis=1, skipna=True)

max_banks = data.idxmax(axis=1, skipna=True)
min_banks = data.idxmin(axis=1, skipna=True)

summary = pd.DataFrame({
    'Average': averages,
    'Max Value': max_values,
    'Max Bank': max_banks,
    'Min Value': min_values,
    'Min Bank': min_banks
})
summary.index = ['2020년','2021년','2022년','2023년','2024년']

average = data.stack().mean()
max = data.stack().max()
min = data.stack().min()

max_bank = data.stack().idxmax()[1]
min_bank = data.stack().idxmin()[1]

all_time_summary = pd.DataFrame({
    'Average': [average],
    'Max Value': [max],
    'Max Bank': [max_bank],
    'Min Value': [min],
    'Min Bank': [min_bank]
})

all_time_summary
```
