```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

file_path = "/home/uk/dev_ws/EDA/data/1.2.data/deposit.xlsx"
deposit = pd.read_excel(file_path)

basic_rate_df = deposit.set_index('은행')
basic_rate_df.rename(columns={'기본금리(단리이자 %)': '예금금리'}, inplace=True)

basic_rate_df = basic_rate_df.dropna(subset=['예금금리'])

plt.figure(figsize=(10, 6))
basic_rate_df['예금금리'].plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('은행별 예금금리', fontsize=16)
plt.ylabel('금리 (%)', fontsize=12)
plt.xlabel('은행', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
